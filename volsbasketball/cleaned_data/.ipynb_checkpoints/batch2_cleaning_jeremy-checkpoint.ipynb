```python
from pathlib import Path

import pandas as pd
```

```python
# read data from Google Sheets clipboard
# NOTE: the first item in the copy/paste will be used as the column number
df = pd.read_clipboard()
df
```

```python
old_identifier_list = ['vols-basketball_1997-outlook',
                       'vols-basketball_1998-outlook',
                       'vols-basketball_1998-postseason',
                       'vols-basketball_1999-postseason',
                       'vols-basketball_2000',
                       'vols-basketball_2000-postseason',
                       'vols-basketball_2001-preseason',
                       'vols-basketball_2001-postseason',
                       'vols-basketball_2002',
                       'vols-basketball_2003',
                       'vols-basketball_2004-preview',
                       'vols-basketball_2004',
                       'vols-basketball_2005',
                       'vols-basketball_2006-postseason',
                       'vols-basketball_2007-postseason',
                       'vols-basketball_2008-postseason',
                       'vols-basketball_2010-postseason',
                       'vols-basketball_2011-postseason',
                      ]
df['old_identifier'] = old_identifier_list
df
```

```python
to_fix_df = pd.DataFrame(columns=['new_identifier', 'old_identifier'])

for old_identifier in old_identifier_list:
    df_row = df.loc[df['old_identifier'] == old_identifier]
    to_fix = df_row.loc[df_row['new_identifier'] != old_identifier]
    if not to_fix.empty:
        to_fix_df = pd.concat([to_fix_df, to_fix], ignore_index=True)
to_fix_df
```

```python
xml_dir_path = Path('/Users/dlisla/Documents/GitHub/athletics/volsbasketball/cleaned_data/batch2/')
```

```python
for index in to_fix_df.index:
    
    old_identifier = to_fix_df['old_identifier'][index]
    xml_name = f'{old_identifier}.xml'
    xml_path = xml_dir_path.joinpath(xml_name)
    if not xml_path.is_file():
        print(f'{xml_path.name} does not exist')
        continue
    print(f'Processing {xml_path.name}')
    
    # get new identifier
    new_identifier = to_fix_df['new_identifier'][index]
    new_xml_name = f'{new_identifier}.xml'
    new_xml_path = xml_dir_path.joinpath(new_xml_name)
    
    # update identifier in xml file
    
    #input file
    with open(xml_path, 'rt') as xml_input:
        # output file
        with open(new_xml_path, 'wt') as xml_output:
            # grab each line
            for line in xml_input:
                # write to output with replaced identifier
                xml_output.write(line.replace(old_identifier, new_identifier))
    if new_xml_path.is_file():
        print(f'{new_xml_path.name} exists, deleting old xml')
        xml_path.unlink()
```
